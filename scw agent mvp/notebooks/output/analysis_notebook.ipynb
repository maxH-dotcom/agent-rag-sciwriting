## 导入库

```python
import pandas as pd
import numpy as np
from linearmodels.panel import PanelOLS
import statsmodels.api as sm
```

## 加载数据

```python
# 加载数据（请确保 CSV 文件与本 notebook 在同一目录）
df = pd.read_csv('zhejiang_carbon.csv')
print(f"数据形状: {df.shape}")
print(f"列名: {list(df.columns)}")
df.head()
```

## 数据预处理

```python
# 检查缺失值
print("缺失值统计:")
print(df.isnull().sum())

# 描述性统计
df[['碳排放总量 10*3t', '农业产值 亿元']].describe()
```

## 固定效应回归

```python
# 设置面板索引
df_panel = df.set_index(['地区', '年份'])

# 准备变量
endog = df_panel['碳排放总量 10*3t']
exog = df_panel[['农业产值 亿元 农药使用量']]

# 添加常数项
exog = sm.add_constant(exog)

# 固定效应回归（within transformation）
model = PanelOLS(endog, exog, entity_effects=True, time_effects=False)
result = model.fit(cov_type='clustered', cluster_entity=True)

print(result.summary)
```
